```python
# Import & Start Fastapi APP
import nest_asyncio
import uvicorn
from fastapi import FastAPI, UploadFile, File
import threading
import time
app = FastAPI()

@app.get("/")
def home():
  return {"message": "Server is running!"}
  nest_asyncio.apply()
```

```python
# Simple OCR Helper
from pdf2image import convert_from_path
import pytesseract

def extract_text_from_pdf(pdf_path):
  full=""
  for p in pdf_path:
    full += pytesseract.image_to_string(p)+"\n"
  return full
```

```python
# Upload API
@app.post("/upload")
async def upload_file(file: UploadFile = File(...)):
  path = f"tmp/{file.filename}"
  with open(path, "wb") as f:
    f.write(file.file.read())
  text = extract_text_from_pdf(path) #OCR
  return{"text": text[:2000]} #return first 2000 chars
```

```python
# Start the server
def run():
  uvicorn.run(app, host="0.0.0.0", port=8000)
  threading.Thread(target=run, daemon=True).start()
  time.sleep(1)
  print("Server is started on port 8000!")
```

```python
# Public URL using Cloudflared
import subprocess, re, time
cmd=["cloudflared", "tunnel", "--url", "http://localhost:8000", "--no-autoupdate"]
proc=subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

public_url=None
start = time.time()
while True:
  line=proc.stdout.readline()
  print(line.strip())

  m=re.search(r"https://trycloudflare.com", line)
  if m:
    public_url=m.group(0)
    break

    if time.time() - start > 25:
      break

    print("\nPUBLIC URL: ", public_url)
    print("open docs here: ", public_url+"/docs")
```
